```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("population_data.csv")
```

```python
data.head()
```

```python
#list of countries for graph
countries = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

#filtering data for selected countries data
filtered_data = data[data['Country Name'].isin(countries)]

#plotting graph
plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel('Countries')
plt.ylabel('Population in 2022')
plt.title('Population Growth')
plt.show()
```
